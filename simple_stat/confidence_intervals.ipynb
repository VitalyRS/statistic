```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from  scipy.stats import norm
%matplotlib inline 


alpha=0.99
a99_1,a99_2=norm.interval(alpha,0,1) 
print norm.interval(alpha,0,1) 
alpha=0.95
a95_1,a95_2=norm.interval(alpha,0,1) 
print norm.interval(alpha,0,1) 
alpha=0.90
a90_1,a90_2=norm.interval(alpha,0,1) 
print norm.interval(alpha,0,1) 
```

```python
x=np.linspace(-3,3,1000)
loc=0
scale=1
standart_samples_norm=norm.pdf(x, loc, scale)
plt.plot(x,standart_samples_norm)
plt.fill_between(x[(x<a99_2)&(x>a99_1)],standart_samples_norm[(x<a99_2)&(x>a99_1)])
plt.title("99%")
plt.figure()
plt.plot(x,standart_samples_norm)
plt.fill_between(x[(x<a95_2)&(x>a95_1)],standart_samples_norm[(x<a95_2)&(x>a95_1)])
plt.title("95%")
plt.figure()
plt.plot(x,standart_samples_norm)
plt.fill_between(x[(x<a90_2)&(x>a90_1)],standart_samples_norm[(x<a90_2)&(x>a90_1)])
plt.title("90%")



plt.show()
```
